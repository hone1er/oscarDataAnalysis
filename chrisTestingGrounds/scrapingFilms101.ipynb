```python
from bs4 import BeautifulSoup
import requests

# URL of page to be scraped
url = 'http://www.films101.com/year10y.htm'

# Retrieve page with the requests module
response = requests.get(url)

# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')

# Examine the results, then determine element that contains sought info
#print(soup.prettify())
```

```python
# results are returned as an iterable list
results = soup.find_all('tr', class_="d")
#results
```

```python
# Create lists to append to
#I did not collect rating as it returned an image and I did not collect Media as a movie's distribution only affects sales

title = []
year = []
rank = []
directors = []
countries = []

# Loop through returned results
for result in results:
    # Error handling
    try:
        # Identify and return title of movie
        t1 = result.find('td', class_="zt1").text
        title.append(t1)

        # Identify and return year of movie
        y1 = result.find('td', class_="zy1").text
        year.append(y1)
        
        # Identify and return rank of movie
        r1 = result.find('td', class_="zn4").text
        rank.append(r1)
        
        # Identify and return directors of movie
        d1 = result.find('td', class_="zd4").text
        directors.append(d1)
        
        # Identify and return country of movie
        c1 = result.find('td', class_="zc4").text
        countries.append(c1)
        
    except AttributeError as e:
        print(e)
```

```python
# Cleaning with Pandas
import pandas as pd

df = pd.DataFrame({
    'Title' : title,
    'Year' : year,
    'Rank' : rank,
    'Directors' : directors,
    'Countries' : countries,
})

df
```

```python
df.to_csv('films101.csv', index=False)
```
